```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('Invistico_Airline.csv')
df.head()
```

```python
# Group by satisfaction and calculate average delays
avg_delays = df.groupby('satisfaction')[['Departure Delay in Minutes', 'Arrival Delay in Minutes']].mean().reset_index()
```

```python
# Bar chart of Average Departure Delay
plt.figure(figsize=(6, 4))
sns.barplot(data=avg_delays, x='satisfaction', y='Departure Delay in Minutes',hue='satisfaction', palette='pastel')
plt.title('Average Departure Delay by Satisfaction')
plt.ylabel('Avg Departure Delay (minutes)')
plt.grid(axis='y')
plt.show()
```

Passengers who are dissatisfied experience longer average departure delay as compared to satisfied.

```python
# Bar chart of Average Arrival Delay
plt.figure(figsize=(6, 4))
sns.barplot(data=avg_delays, x='satisfaction', y='Arrival Delay in Minutes',hue='satisfaction', palette='Set2')
plt.title('Average Arrival Delay by Satisfaction')
plt.ylabel('Avg Arrival Delay (minutes)')
plt.grid(axis='y')
plt.show()
```

Passengers who are dissatisfied experience longer average arrival delay of 18 min as compared to satisfied(12.5).
